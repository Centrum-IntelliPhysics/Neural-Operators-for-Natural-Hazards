```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat, savemat
import sys

# ux
fno_ux = loadmat('fno/ux/Results/WindResponse_test.mat') # Load the .mat file
targets_ux = fno_ux['y1_test']
targets_ux_max = np.max(targets_ux, axis=1)
deeponet_ux = loadmat('deeponet+fno/ux/Results/pred_deeponet_test.mat') # Load the .mat file
deeponet_ux_max = np.max(deeponet_ux['disp_pred'], axis=1).reshape(-1,1)
deeponet_fno_ux = loadmat('deeponet+fno/ux/Results/Wind_responses_test_fno.mat') # Load the .mat file
deeponet_fno_ux_max = np.max(deeponet_fno_ux['y_pred'], axis=1)
max_ux = np.concatenate((targets_ux_max, deeponet_ux_max, deeponet_fno_ux_max), axis=-1)

labels = ['Target', 'DeepONet', 'DeepONet+FNO']
colors = ['black', 'royalblue', 'magenta']

plt.figure(figsize=(8, 5))
bins = 20  

for i in range(3):
    plt.hist(max_ux[:, i], bins=bins, alpha=0.5, label=labels[i], color=colors[i], edgecolor='k')

plt.xlabel('Maximum $u_x$ Response')
plt.ylabel('Frequency')
#plt.title('Histogram of Maximum $u_x$ Displacement')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('maxvalue_plots/histogram_max_ux_comparison.png', dpi=300)
plt.show()

plt.figure(figsize=(6, 5))
plt.boxplot([max_ux[:, 0], max_ux[:, 1].flatten(), max_ux[:, 2]], 
            labels=labels,
            patch_artist=True,
            boxprops=dict(facecolor='lightgray', color='k'),
            medianprops=dict(color='red'))

plt.ylabel('Maximum $u_x$ Response')
#plt.title('Maximum $u_x$ Displacement')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('maxvalue_plots/boxplot_max_ux_comparison.png', dpi=300)
plt.show()
```
