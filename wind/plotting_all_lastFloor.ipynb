```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat, savemat
import sys

def calculate_mse(y_true, y_pred):
    y_true = y_true.reshape(y_true.shape[0], -1)
    y_pred = y_pred.reshape(y_pred.shape[0], -1)
    return np.mean((np.array(y_true) - np.array(y_pred)) ** 2, axis = -1)

def calculate_l2_error(y_true, y_pred):
    y_true = y_true.reshape(y_true.shape[0], -1)
    y_pred = y_pred.reshape(y_pred.shape[0], -1)
    relative_errors = np.linalg.norm(y_true - y_pred, axis=0) / (np.linalg.norm(y_true, axis=0) + np.finfo(float).eps)
    return np.mean(relative_errors)

def calculate_max_abs(y_true, y_pred):
    y_true = y_true.reshape(y_true.shape[0], -1)
    y_pred = y_pred.reshape(y_pred.shape[0], -1)
    abs_errors = np.abs(y_true - y_pred)
    n_samples = y_true.shape[0]
    max_abs_errors = np.max(abs_errors, axis=0)
    # Error Variance (average variance across samples)
    variances = [np.var(abs_errors[i]) for i in range(n_samples)]
    avg_variance = np.mean(variances)
    return max_abs_errors, avg_variance

errors = np.zeros((12, 4))
fno_ux = loadmat('fno/ux/Results/WindResponse_test.mat') # Load the .mat file
targets_ux = fno_ux['y1_test']
pred_ux_fno = fno_ux['y1_pred']
mse_ux_fno_samp = calculate_mse(targets_ux,pred_ux_fno)
mse_ux_fno = np.mean(mse_ux_fno_samp)
l2_ux_fno = calculate_l2_error(targets_ux,pred_ux_fno)
a,b = calculate_max_abs(targets_ux, pred_ux_fno)
print("Max absolute error:", np.max(a))
print("Mean Max absolute error:", np.mean(a))
print("Average variance of absolute error:", b)

# errors[0,0] = mse_ux_fno
# errors[0,1] = np.min(mse_ux_fno_samp)
# min_index = np.argmin(mse_ux_fno_samp)
# print("Index of minimum ux FNO:", min_index)
# errors[0,2] = np.max(mse_ux_fno_samp)
# errors[0,3] = l2_ux_fno

sa_fno_ux = loadmat('sa-fno/ux/Results/WindResponse_test.mat') # Load the .mat file
pred_ux_sa_fno = sa_fno_ux['y1_pred']
mse_ux_sa_fno_samp = calculate_mse(targets_ux,pred_ux_sa_fno)
mse_ux_sa_fno = np.mean(mse_ux_sa_fno_samp)
l2_ux_sa_fno = calculate_l2_error(targets_ux,pred_ux_sa_fno)
a,b = calculate_max_abs(targets_ux, pred_ux_sa_fno)
print("Max absolute error:", np.max(a))
print("Mean Max absolute error:", np.mean(a))
print("Average variance of absolute error:", b)


# errors[1,0] = mse_ux_sa_fno
# errors[1,1] = np.min(mse_ux_sa_fno_samp)
# min_index = np.argmin(mse_ux_sa_fno_samp)
# print("Index of minimum ux SA-FNO:", min_index)
# errors[1,2] = np.max(mse_ux_sa_fno_samp)
# errors[1,3] = l2_ux_sa_fno

deeponet_fno_ux = loadmat('deeponet+fno/ux/Results/Wind_responses_test_fno.mat') # Load the .mat file
pred_ux_deeponet_fno = deeponet_fno_ux['y_pred']
mse_ux_deeponet_fno_samp = calculate_mse(targets_ux,pred_ux_deeponet_fno)
mse_ux_deeponet_fno = np.mean(mse_ux_deeponet_fno_samp)
l2_ux_deeponet_fno = calculate_l2_error(targets_ux,pred_ux_deeponet_fno)

# errors[2,0] = mse_ux_deeponet_fno
# errors[2,1] = np.min(mse_ux_deeponet_fno_samp)
# min_index = np.argmin(mse_ux_deeponet_fno_samp)
# print("Index of minimum ux DeepONet-FNO:", min_index)
# errors[2,2] = np.max(mse_ux_deeponet_fno_samp)
# errors[2,3] = l2_ux_deeponet_fno

deeponet_ux = loadmat('deeponet+fno/ux/Results/pred_deeponet_test.mat') # Load the .mat file
pred_ux_deeponet = deeponet_ux['disp_pred']
mse_ux_deeponet_samp = calculate_mse(targets_ux,pred_ux_deeponet)
mse_ux_deeponet = np.mean(mse_ux_deeponet_samp)
l2_ux_deeponet = calculate_l2_error(targets_ux,pred_ux_deeponet)

# errors[3,0] = mse_ux_deeponet
# errors[3,1] = np.min(mse_ux_deeponet_samp)
# min_index = np.argmin(mse_ux_deeponet_samp)
# print("Index of minimum ux DeepONet:", min_index)
# errors[3,2] = np.max(mse_ux_deeponet_samp)
# errors[3,3] = l2_ux_deeponet

fno_uy = loadmat('fno/uy/Results/WindResponse_test.mat') # Load the .mat file
targets_uy = fno_uy['y1_test']
pred_uy_fno = fno_uy['y1_pred']
mse_uy_fno_samp = calculate_mse(targets_uy,pred_uy_fno)
mse_uy_fno = np.mean(mse_uy_fno_samp)
l2_uy_fno = calculate_l2_error(targets_uy,pred_uy_fno)
a,b = calculate_max_abs(targets_uy, pred_uy_fno)
print("Max absolute error:", np.max(a))
print("Mean Max absolute error:", np.mean(a))
print("Average variance of absolute error:", b)

# errors[4,0] = mse_uy_fno
# errors[4,1] = np.min(mse_uy_fno_samp)
# min_index = np.argmin(mse_uy_fno_samp)
# print("Index of minimum uy FNO:", min_index)
# errors[4,2] = np.max(mse_uy_fno_samp)
# errors[4,3] = l2_uy_fno

sa_fno_uy = loadmat('sa-fno/uy/Results/WindResponse_test.mat') # Load the .mat file
pred_uy_sa_fno = sa_fno_uy['y1_pred']
mse_uy_sa_fno_samp = calculate_mse(targets_uy,pred_uy_sa_fno)
mse_uy_sa_fno = np.mean(mse_uy_sa_fno_samp)
l2_uy_sa_fno = calculate_l2_error(targets_uy,pred_uy_sa_fno)
a,b = calculate_max_abs(targets_uy, pred_uy_sa_fno)
print("Max absolute error:", np.max(a))
print("Mean Max absolute error:", np.mean(a))
print("Average variance of absolute error:", b)

# errors[5,0] = mse_uy_sa_fno
# errors[5,1] = np.min(mse_uy_sa_fno_samp)
# min_index = np.argmin(mse_uy_sa_fno_samp)
# print("Index of minimum uy SA-FNO:", min_index)
# errors[5,2] = np.max(mse_uy_sa_fno_samp)
# errors[5,3] = l2_uy_sa_fno

deeponet_fno_uy = loadmat('deeponet+fno/uy/Results/Wind_responses_test_fno.mat') # Load the .mat file
pred_uy_deeponet_fno = deeponet_fno_uy['y_pred'] 
mse_uy_deeponet_fno_samp = calculate_mse(targets_ux,pred_uy_deeponet_fno)
mse_uy_deeponet_fno = np.mean(mse_uy_deeponet_fno_samp)
l2_uy_deeponet_fno = calculate_l2_error(targets_uy,pred_uy_deeponet_fno)

# errors[6,0] = mse_uy_deeponet_fno
# errors[6,1] = np.min(mse_uy_deeponet_fno_samp)
# min_index = np.argmin(mse_uy_deeponet_fno_samp)
# print("Index of minimum uy DeepONet-FNO:", min_index)
# errors[6,2] = np.max(mse_uy_deeponet_fno_samp)
# errors[6,3] = l2_uy_deeponet_fno

deeponet_uy = loadmat('deeponet+fno/uy/Results/pred_deeponet_test.mat') # Load the .mat file
pred_uy_deeponet = deeponet_uy['disp_pred']
mse_uy_deeponet_samp = calculate_mse(targets_uy,pred_uy_deeponet)
mse_uy_deeponet = np.mean(mse_uy_deeponet_samp)
l2_uy_deeponet = calculate_l2_error(targets_uy,pred_uy_deeponet)

errors[7,0] = mse_uy_deeponet
errors[7,1] = np.min(mse_uy_deeponet_samp)
min_index = np.argmin(mse_uy_deeponet_samp)
print("Index of minimum uy DeepONet:", min_index)
errors[7,2] = np.max(mse_uy_deeponet_samp)
errors[7,3] = l2_uy_deeponet

fno_uz = loadmat('fno/uz/Results/WindResponse_test.mat') # Load the .mat file
targets_uz = fno_uz['y1_test']
pred_uz_fno = fno_uz['y1_pred']
mse_uz_fno_samp = calculate_mse(targets_uz,pred_uz_fno)
mse_uz_fno = np.mean(mse_uz_fno_samp)
l2_uz_fno = calculate_l2_error(targets_uz,pred_uz_fno)
a,b = calculate_max_abs(targets_uz, pred_uz_fno)
print("Max absolute error:", np.max(a))
print("Mean Max absolute error:", np.mean(a))
print("Average variance of absolute error:", b)

# errors[8,0] = mse_uz_fno
# errors[8,1] = np.min(mse_uz_fno_samp)
# min_index = np.argmin(mse_uz_fno_samp)
# print("Index of minimum uz FNO:", min_index)
# errors[8,2] = np.max(mse_uz_fno_samp)
# errors[8,3] = l2_uz_fno

sa_fno_uz = loadmat('sa-fno/uz/Results/WindResponse_test.mat') # Load the .mat file
pred_uz_sa_fno = sa_fno_uz['y1_pred']
mse_uz_sa_fno_samp = calculate_mse(targets_uz,pred_uz_sa_fno)
mse_uz_sa_fno = np.mean(mse_uz_sa_fno_samp)
l2_uz_sa_fno = calculate_l2_error(targets_uz,pred_uz_sa_fno)
a,b = calculate_max_abs(targets_uz, pred_uz_sa_fno)
print("Max absolute error:", np.max(a))
print("Mean Max absolute error:", np.mean(a))
print("Average variance of absolute error:", b)

# errors[9,0] = mse_uz_sa_fno
# errors[9,1] = np.min(mse_uz_sa_fno_samp)
# min_index = np.argmin(mse_uz_sa_fno_samp)
# print("Index of minimum uz SA-FNO:", min_index)
# errors[9,2] = np.max(mse_uz_sa_fno_samp)
# errors[9,3] = l2_uz_sa_fno

deeponet_fno_uz = loadmat('deeponet+fno/uz/Results/Wind_responses_test_fno.mat') # Load the .mat file
pred_uz_deeponet_fno = deeponet_fno_uz['y_pred']
mse_uz_deeponet_fno_samp = calculate_mse(targets_uz,pred_uz_deeponet_fno)
mse_uz_deeponet_fno = np.mean(mse_uz_deeponet_fno_samp)
l2_uz_deeponet_fno = calculate_l2_error(targets_uz,pred_uz_deeponet_fno)

# errors[10,0] = mse_uz_deeponet_fno
# errors[10,1] = np.min(mse_uz_deeponet_fno_samp)
# min_index = np.argmin(mse_uz_deeponet_fno_samp)
# print("Index of minimum uy DeepONet-FNO:", min_index)
# errors[10,2] = np.max(mse_uz_deeponet_fno_samp)
# errors[10,3] = l2_uz_deeponet_fno

deeponet_uz = loadmat('deeponet+fno/uz/Results/pred_deeponet_test.mat') # Load the .mat file
pred_uz_deeponet = deeponet_uz['disp_pred']
mse_uz_deeponet_samp = calculate_mse(targets_uz,pred_uz_deeponet)
mse_uz_deeponet = np.mean(mse_uz_deeponet_samp)
l2_uz_deeponet = calculate_l2_error(targets_uz,pred_uz_deeponet)

# errors[11,0] = mse_uz_deeponet
# errors[11,1] = np.min(mse_uz_deeponet_samp)
# min_index = np.argmin(mse_uz_deeponet_samp)
# print("Index of minimum uy DeepONet:", min_index)
# errors[11,2] = np.max(mse_uz_deeponet_samp)
# errors[11,3] = l2_uz_deeponet

# print(errors)
# print(pred_fno_ux.shape)
```
